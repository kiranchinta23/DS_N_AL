# <u>Double Linked list Node</u>
<img src='d_node.png'>

```python
class Node:
    __slots__='element','next','prev'
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev
```

```python
n1=Node(4,None,None)
n2=Node(5,None,None)
```

```python
print(n1.element)
print(n1.next)
print(n1.prev)
```

# <u>Display</u>
<img src='d_display.png'>

```python
class Node:
    __slots__='element','next','prev'
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev
        
class dLL:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
    
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
            newest.prev=self.tail
        self.tail=newest
        self.size+=1
    
    def display(self):
        p=self.head
        while p:
            print(p.element,end='<-->')
            p=p.next
        print()
        
    def displayrev(self):
        p=self.tail
        while p:
            print(p.element,end='<-->')
            p=p.prev
        print()
            
```

```python
d=dLL()
d.addlast(10)
d.display()
d.addlast(20)
print('size is:',len(d))
d.display()
d.addlast(40)
d.addlast(50)
d.addlast(60)
d.displayrev()
```

# <u>Addany</u>

```python
class Node:
    __slots__='element','next','prev'
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev
        
class dLL:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
    
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
            newest.prev=self.tail
        self.tail=newest
        self.size+=1
    
    def addfirst(self,value):
        newest=Node(value,None,None)
        if self.isempty():
            self.tail=newest
        else:
            newest.next=self.head
            self.head.prev=newest
        self.head=newest
        self.size+=1
    
    def addany(self,value,pos):
                
        if pos <=0 or pos>len(self):
            print('can not insert outside of size')
            return
        
        elif pos==1:
            self.addfirst(value)
            return  
        
        newest=Node(value,None,None)
        p=self.head
        i=1
        while i<pos-1:
            p=p.next
            i=i+1
        newest.next=p.next
        p.next.prev=newest
        p.next=newest
        newest.prev=p
        self.size+=1
        
       
    
    def display(self):
        p=self.head
        while p:
            print(p.element,end='<-->')
            p=p.next
        print()
        
    def displayrev(self):
        p=self.tail
        while p:
            print(p.element,end='<-->')
            p=p.prev
        print()
            
```

# <u>RemoveAny</u>

```python
class Node:
    __slots__='element','next','prev'
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev
        
class dLL:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
    
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
            newest.prev=self.tail
        self.tail=newest
        self.size+=1
    
    def addfirst(self,value):
        newest=Node(value,None,None)
        if self.isempty():
            self.tail=newest
        else:
            newest.next=self.head
            self.head.prev=newest
        self.head=newest
        self.size+=1
    
    def addany(self,value,pos):
                
        if pos <1 or pos>len(self):
            print('can not insert outside of size')
            return
        
        elif pos==1:
            self.addfirst(value)
            return  
        
        newest=Node(value,None,None)
        p=self.head
        i=1
        while i<pos-1:
            p=p.next
            i=i+1
        newest.next=p.next
        p.next.prev=newest
        p.next=newest
        newest.prev=p
        self.size+=1
        
       
    
    def display(self):
        p=self.head
        while p:
            print(p.element,end='<-->')
            p=p.next
        print()
        
    def displayrev(self):
        p=self.tail
        while p:
            print(p.element,end='<-->')
            p=p.prev
        print()
    
    def removefirst(self):
        if self.isempty():
            print('list is empty')
            return
        
        elif self.size==1:
            e=self.head.element
            self.head=None
            self.tail=None
            self.size-=1
            return e
        
        e=self.head.element
        self.head=self.head.next
        self.head.prev=None
        self.size-=1
        return e
    
    def removelast(self):
        if self.isempty():
            print('list is empty')
            return
        
        elif self.size==1:
            e=self.head.element
            self.head=None
            self.tail=None
            self.size-=1
            return e
        
        e=self.tail.element
        self.tail=self.tail.prev
        self.tail.next=None
        self.size-=1
        return e
    
    def removeany(self,pos):
        if pos <=1 or pos>=len(self):
            print('can not remove outside of size')
            return
                
        p=self.head
        i=1
        while i<pos-1:
            p=p.next
            i=i+1
        e=p.next.element
        p.next=p.next.next
        p.next.prev=p
        self.size-=1
        return e
```
