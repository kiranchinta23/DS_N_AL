# Linked List

```python
class Node:
    __slots__='element','next'
    def __init__(self,element,next):
        self.element=element
        self.next=next
        
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
        self.tail=newest
        self.size+=1
        
    def display(self):
        p=self.head
        while p:
            print(p.element,end='-->')
            p=p.next
        print()
```

```python
L=LinkedList()
L.addlast(7)
L.addlast(3)
L.addlast(10)
print('length of ll is:',len(L))
L.display()
L.addlast(1)
L.addlast(3)
print('length of ll is:',len(L))
L.display()
```

# <u>Searching</u>

<img src='searching.png'>

```python
class Node:
    __slots__='element','next'
    def __init__(self,element,next):
        self.element=element
        self.next=next
        
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
        self.tail=newest
        self.size+=1
        
    def display(self):
        p=self.head
        while p:
            print(p.element,end='-->')
            p=p.next
        print()
        
    def search(self,key):
        p=self.head
        index=0
        while p:
            if p.element==key:
                return index
            p=p.next
            index=index+1
        return -1
```

```python
class Node:
    __slots__='element','next'
    def __init__(self,element,next):
        self.element=element
        self.next=next
        
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None)
        if self.isempty():
            self.head=newest
        else:
            self.tail.next=newest
        self.tail=newest
        self.size+=1
    
    def addfirst(self,value):
        newest=Node(value,None)
        if self.isempty():
            self.head=newest
            self.tail=newest
        else:
            newest.next=self.head
        self.head=newest
        self.size+=1
    
    def addany(self,value,pos):
        if pos <=0 or pos>len(self):
            print('can not insert outside of size')
            return
        
        if pos==1:
            self.addfirst(value)
            return
            
        newest=Node(value,None)
        p=self.head
        i=1
        while i<pos-1:
            p=p.next
            i=i+1
        newest.next=p.next
        p.next=newest
        self.size+=1
        
        
        
    def display(self):
        p=self.head
        while p:
            print(p.element,end='-->')
            p=p.next
        print()
        
    def search(self,key):
        p=self.head
        index=0
        while p:
            if p.element==key:
                return index
            p=p.next
            index=index+1
        return -1

    def removefirst(self):
        if self.isempty():
            print('list is empty')
            return
        elif self.size==1:
            e=self.head.element
            self.head=None
            self.tail=None
            return e
        e=self.head.element
        self.head=self.head.next
        self.size-=1
        return e
    
    def removelast(self):
        if self.isempty():
            print('list is empty')
            return
        elif self.size==1:
            e=self.head.element
            self.head=None
            self.tail=None
            return e
        p=self.head
        i=1 
        while i< len(self)-1:
            p=p.next
            i+=1
        e=self.tail.element
        self.tail.next=None
        self.tail=p
        self.size-=1
        return e
    
    def removeany(self,pos):
        if pos <=0 or pos>len(self):
            print('can not remove outside of size')
            return
        elif pos==1:
            self.removefirst()
            return
        
        p=self.head
        i=1
        while i<pos-1:
            p=p.next
            i=i+1
        e=p.next.element
        p.next=p.next.next
        self.size-=1
        return e
```
