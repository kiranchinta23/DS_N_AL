<img src='c_display.png'>

```python
class Node:
    
    def __init__(self,element,next):
        self.element=element
        self.next=next

class CircularLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size==0
    
    def addlast(self,value):
        newest=Node(value,None)
        if self.isempty():
            self.head=newest
            newest.next=newest
        else:
            newest.next=self.tail.next
            self.tail.next=newest
        self.tail=newest
        self.size+=1
    
    def display(self):
        p=self.head
        i=0
        while i<len(self):
            print(p.element,end='-->')
            p=p.next
            i+=1
        print()
    
    def search(self,key):
        p=self.head
        i=0
        while i<len(self):
            if p.element==key:
                print('element found at index: ',i)
                return
            p=p.next
            i=i+1
        return -1
        
        
```

```python
C = CircularLinkedList()
C.addlast(7)
C.addlast(4)
C.addlast(12)
C.display()
print('Size:',len(C))
C.addlast(8)
C.addlast(3)
C.search(12)
C.search(1)
```
